En primera instancia, mencionamos a python la utilización de las siguientes librerías: pandas, numpy, yfinance y plotly.

------

```python
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

Luego hacemos el llamado al Ticker respectivo que esta alojado en los servidores de Yahoo! finance

-----------

```python
qubt = yf.Ticker("QUBT")
```

Realizado lo anterior, procedemos a extraer la data histórica de QUBT de Yahoo! finance

---------------

```python
qubt_historial = qubt.history(period = "5y")
```

Luego de realizar la extracción de la data histórica de qubt, procedemos a restaurar el index convencional, dado que la fecha esta tomando ese papel en la data histórica

--------------

```python
qubt_historial.reset_index(inplace = True)
qubt_historial.head()
```

Luego de realizar la restauración del index, procedemos a realizar la función para poder gráficar la volatilidad de los datos, al último cierre guardado

```python
def make_graph(stock_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price"))
    stock_data_specific = stock_data[stock_data.Date >= '2020-01-09']
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data_specific.Date, infer_datetime_format=True), y=stock_data_specific.Close.astype("float"), name="Share Price"), row=1, col=1)
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_layout(showlegend=False,
    height=900,
    title=stock,
    xaxis_rangeslider_visible=True)
    fig.show()
    
```

Ejecutamos la función con la data extraída

```python
make_graph(qubt_historial, "QUBT")
```
